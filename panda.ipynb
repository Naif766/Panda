```python
import pandas as pd
```

```python
df = pd.read_csv('motodata.csv')

print(df.to_string()) 
```

```python
df["Selling_Price"].fillna(df["Selling_Price"].mode()[0], inplace = True)
df
```

```python
df["Selling_Price"].dropna()
```

```python
for i in df.index:

    if df.loc[i , "Selling_Price"]>100:
        df.loc[i , "Selling_Price"] = df["Selling_Price"].mode()[0]
```

```python
df.drop_duplicates(inplace = True)
df
```
